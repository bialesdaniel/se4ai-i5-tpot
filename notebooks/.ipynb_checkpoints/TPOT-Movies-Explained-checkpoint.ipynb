```python
import pandas as pd

movies_raw = pd.read_csv('../movie-data.csv')
users_raw = pd.read_csv('../users-data.csv')
ratings_raw = pd.read_csv('../ratings-data.csv')

users_raw.set_index('user_id', inplace=True)
movies_raw.set_index('movie_id', inplace=True)
```

```python
movies_raw.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
import json

ratings = ratings_raw.join(users_raw, on='user_id', how='left')
ratings = ratings.join(movies_raw, on='movie_id', how='left')

ratings.loc[ratings['genres'].isnull(),['genres']] = ratings.loc[ratings['genres'].isnull(),'genres'].apply(lambda x: '[]')
ratings['genres'] = ratings['genres'].apply(lambda x: np.array(x[1:len(x)-1].replace("'","").replace(" ","").split(',')))
#ratings['genres'] = ratings['genres'].apply(lambda x: tuple(map(tuple, x)))

mlb = MultiLabelBinarizer()
mlb.fit(ratings['genres'])
print(mlb.classes_)
ratings = ratings.join(pd.DataFrame(mlb.transform(ratings['genres']),columns=mlb.classes_))
ratings.head()
```

```python
# ratings.loc[ratings['production_companies'].isnull(),['production_companies']] = ratings.loc[ratings['production_companies'].isnull(),'production_companies'].apply(lambda x: '[]')
# ratings['production_companies'] = ratings['production_companies'].apply(lambda x: np.array(x[1:len(x)-1].replace("'","").replace(" ","").split(',')))

# mlb = MultiLabelBinarizer()
# mlb.fit(ratings['production_companies'])
# print(mlb.classes_)
# ratings = ratings.join(pd.DataFrame(mlb.transform(ratings['production_companies']),columns=mlb.classes_))
# ratings.head()
```

```python
ratings.loc[ratings['production_countries'].isnull(),['production_countries']] = ratings.loc[ratings['production_countries'].isnull(),'production_countries'].apply(lambda x: '[]')
ratings['production_countries'] = ratings['production_countries'].apply(lambda x: np.array(x[1:len(x)-1].replace("'","").replace(" ","").split(',')))

mlb = MultiLabelBinarizer()
mlb.fit(ratings['production_countries'])
print(mlb.classes_)
ratings = pd.concat([ratings,pd.DataFrame(mlb.transform(ratings['production_countries']),columns=mlb.classes_)],axis=1)
ratings.head()
```

```python
ratings.loc[ratings['spoken_languages'].isnull(),['spoken_languages']] = ratings.loc[ratings['spoken_languages'].isnull(),'spoken_languages'].apply(lambda x: '[]')
ratings['spoken_languages'] = ratings['spoken_languages'].apply(lambda x: np.array(x[1:len(x)-1].replace("'","").replace(" ","").split(',')))

mlb = MultiLabelBinarizer()
mlb.fit(ratings['spoken_languages'])
print(mlb.classes_)
ratings = pd.concat([ratings,pd.DataFrame(mlb.transform(ratings['spoken_languages']),columns=mlb.classes_)],axis=1)
ratings.head()
```

```python
mlb = MultiLabelBinarizer()
mlb.fit(ratings['gender'])
print(mlb.classes_)
ratings = pd.concat([ratings,pd.DataFrame(mlb.transform(ratings['gender']),columns=mlb.classes_)],axis=1)
ratings.head()
```

```python
mlb = MultiLabelBinarizer()
mlb.fit([ratings['occupation']])
print(mlb.classes_)
ratings = pd.concat([ratings,pd.DataFrame(mlb.transform(ratings['occupation']),columns=mlb.classes_)],axis=1)
ratings.head()
```

```python
ratings = ratings.dropna(subset=['release_date'])
ratings['year'] = ratings['release_date'].apply(lambda x: str(x)[:4])
ratings['month'] = ratings['release_date'].apply(lambda x: str(x)[5:7])
ratings = ratings.astype({'year':'int64', 'month':'int64'})
ratings.head()
```
